```python
import os
from tqdm import tqdm
import json
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from IPython.display import Image,display
np.random.seed(123)
```

```python
def split_name(name):
    study_id = name.split('.')[0]
    date = name.split()[1]
    split_date =date.split('-')
    year = split_date[0]
    month = split_date[1]
    day = split_date[2]
    month = month.lstrip("0")
    day = day.lstrip("0")
    date = month + "/" + day + "/" + year
    
    return study_id,date
```

```python
food_labels = pd.read_csv("food_labels_copy.csv") 
food_labels = food_labels[["StudyID","Date","Unnamed: 9"]]
food_labels = food_labels.rename(columns={"Unnamed: 9":"fine labels"})
```

```python
# Pair images from directory with labels from csv file
im_directory = './Food_photos_copy'
image_names = os.listdir(im_directory)

data = pd.DataFrame(columns=['name', 'fine labels'])
for name in tqdm(image_names):
    study_id,date = split_name(name)
    rows = food_labels.loc[(food_labels['StudyID'] == study_id) & (food_labels['Date'] == date)]
    fine_labels_rows = rows["fine labels"].tolist()
    data = data.append({'name': name, 'fine labels': fine_labels_rows}, ignore_index=True)
```

```python
# Show a few images with corresponding labels
for name in image_names[:10]:
    study_id,date = split_name(name)
    rows = food_labels.loc[(food_labels['StudyID'] == study_id) & (food_labels['Date'] == date)]
    labels = rows["fine labels"].tolist()
    
    display(Image(filename=os.path.join(im_directory,name)))
    print(labels)
```

```python
# Now you have a dataframe w/ filenames and fine labels
# Next: convert to coarse labels via matrix in the fine_to_coarse.csv
# add a coarse label column to the dataframe
```

```python
fine_to_coarse = pd.read_csv("fine_to_coarse.csv",encoding = "utf-8") 
```

```python
fine_to_coarse.head()
```

```python
fine_labels = fine_to_coarse.columns[1:]
coarse_labels = fine_to_coarse["Coarse names"]
```

```python
data.head()
```

```python
# Fill coarse label column in dataframe
for index,row in tqdm(data.iterrows()):
    fine_labels_row = row["fine labels"]
    data["coarse labels"].iloc[index] = set()
    for fine_label in fine_labels_row:
        coarse_index = np.where(fine_to_coarse[fine_label])
        coarse_label = coarse_labels.iloc[coarse_index].to_numpy()
        if(not coarse_label): continue
        coarse_label = coarse_label[0]
        if(not data["coarse labels"].iloc[index]):
            data["coarse labels"].iloc[index] = set()
        data["coarse labels"].iloc[index].add(coarse_label)
```

```python
# not all of the coarse labels are lists, make sure converted to list!
data['coarse labels'] = data['coarse labels'].apply(list)
```

```python
types = data['coarse labels'].apply(type)
types.to_numpy()
prob = np.where(types == type(None))
```

```python
cond = [len(data['coarse labels'].iloc[i])==0 for i in range(len(data))]
data.loc[cond]
```

```python
#there are 121 images with no labels, remove them
data_clean = data.drop(np.where(cond)[0],axis=0)
```

```python
# dropped empty rows!
# check that there are no empty rows now
cond = [len(data_clean['coarse labels'].iloc[i])==0 for i in range(len(data_clean))]
np.where(cond)
```

```python
# data is clean, now save the pandas dataframe and start another notebook for deep learning
data_clean.to_pickle("./data.pkl")
```
